```python
import pandas as pd

fecha = '31/12/2021'
f_val = pd.to_datetime(fecha, format='%d/%m/%Y')

s = pd.date_range(start=f_val,periods=15,freq='M')
s
```

```python
df = pd.DataFrame(s, columns=['Fecha'])
df
```

```python
df['Periodo'] = df['Fecha'].dt.to_period(freq='Q')
df
```

```python
df['Año'] = df['Fecha'].dt.year
df
```

```python
df['N_Dias'] = df['Fecha'].diff()
df
```
