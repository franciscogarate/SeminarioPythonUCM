```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
x = (1, 2, 3, 4, 5)
y = (1.25, 5, 11.25, 20, 30.5)

st.linregress(x,y)
```

```python
plt.scatter(x,y)
```

```python
x_recta = np.linspace(1,5,20)
y_recta = -8.45 + x_recta * 7.35
plt.scatter(x,y)
plt.plot(x_recta,y_recta, color='r')
```
