# Pandas y numpy

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(2021)
datos = np.random.randn(10,1)
datos
```

```python
df = pd.DataFrame(datos, columns=['A'])
df
```

```python
df['B'] = df['A'].abs()
df
```

```python
df['C'] = df['A'] + df['B']
df
```

```python
df['D'] = df['C'] + 100
df
```

```python
df['Diff'] = df['D'].diff().fillna(0)
df
```

```python
df['Diff'].plot()
```

```python
#df['A_mean'] = df['A'].sum() / df['A'].count()
df['A_mean'] = df['A'].mean()
df
```

```python
df['C_sqrt'] = np.sqrt(df['C'])
df
```
