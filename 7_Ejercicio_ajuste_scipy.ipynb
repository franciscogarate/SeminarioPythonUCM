```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('siniestros.xls')
df.head()
```

```python
model = st.norm.fit(df['Importe'])
model
```

```python
media, desv = model
media
```

```python
desv
```

```python
x = np.linspace(0, 35,100)
y = st.norm.pdf(x, media, desv)
plt.plot(x,y)
plt.hist(df['Importe'], bins=15, density=True)
```
