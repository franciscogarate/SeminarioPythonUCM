```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# data = https://vincentarelbundock.github.io/Rdatasets/doc/AER/USAirlines.html
data = pd.read_csv('airlines.csv', index_col=0)
data.head()
```

```python
formula='cost ~ output + pf + lf'

model = smf.glm(formula=formula, data=data, family=sm.families.Poisson())
result = model.fit()
print(result.summary())
```

```python
data['Estimación'] = result.predict(data)
data['Estimación'] = data['Estimación'].round(0)
data['Error'] = data['cost'] - data['Estimación']
data.head()
```

```python
import numpy as np
data['Error2'] = ((data['cost'] - data['Estimación']) ** 2).round(2)
np.sqrt(data['Error2'].mean())
```
